### We have missing values in Income column

```python
if 'Income' in df.columns :
    df['Income']=df['Income'].fillna(df['Income'].median())
    print(df.isnull().sum())
    
```

```python
df.dropna() # for any missing values in other columns
```

```python
# Standardize the Categorical Columns
cat_cols=df.select_dtypes(include='object').columns
cat_cols
```

```python
df['Education']=df['Education'].str.strip().str.lower()
```

```python
df['Marital_Status']=df['Marital_Status'].str.strip().str.lower()
```

```python
df.head()
```

```python
# Convert the Dt_Customer  column dtype into datetime format
df['Dt_Customer']=pd.to_datetime(df['Dt_Customer'],errors='coerce')
display(df.head())
print("\nDt_customer column datatype after conversion : ",df['Dt_Customer'].dtype)
```

```python
df.columns=[cols.strip().lower().replace(" ","_").replace("-","_") for cols in df.columns]
```
